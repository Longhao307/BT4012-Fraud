```python
import pandas as pd

# Load the CSV files
file_path1 = "Fraudulent_E-Commerce_Transaction_Data.csv"
file_path2 = "Fraudulent_E-Commerce_Transaction_Data_2.csv"

# Read the CSV files
data1 = pd.read_csv(file_path1)
data2 = pd.read_csv(file_path2)

# Show the head of each dataframe
head1 = data1.head()
head2 = data2.head()

head1, head2
```

```python
data1
```

```python
# Combine the two dataframes
combined_data = pd.concat([data1, data2])

# Save the combined dataframe to a CSV file
combined_data.to_csv('Combined_Fraudulent_Transactions.csv', index=False)
```

```python
# Load the recently saved combined CSV to count the number of fraudulent entries
combined_data_loaded = pd.read_csv('Combined_Fraudulent_Transactions.csv')
combined_data_loaded
```

```python

# Count the number of entries where 'Is Fraudulent' is 1
fraudulent_count = combined_data_loaded[combined_data_loaded['Is Fraudulent'] == 1].shape[0]
fraudulent_count
```

```python
import matplotlib.pyplot as plt

# Count the instances of fraudulent and non-fraudulent transactions
fraud_counts = combined_data_loaded['Is Fraudulent'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 5))
fraud_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Distribution of Fraudulent Transactions')
plt.xlabel('Is Fraudulent')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Non-Fraudulent', 'Fraudulent'], rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Convert 'Transaction Date' to datetime format
combined_data_loaded['Transaction Date'] = pd.to_datetime(combined_data_loaded['Transaction Date'])

# Find the minimum and maximum transaction dates
date_min = combined_data_loaded['Transaction Date'].min()
date_max = combined_data_loaded['Transaction Date'].max()

date_min, date_max
```
